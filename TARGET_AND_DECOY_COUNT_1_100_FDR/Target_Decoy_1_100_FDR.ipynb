```python
import os
import pandas as pd
import glob

# get a list of all csv files in a directory
csv_files = glob.glob('/Users/anabelyong/miniconda3/lib/python3.10/site-packages/jupyterlab/Subset750RawfilesCut/Conversion_Delete/*.csv')

# create an empty list to store the results for each file
results_list = []

# loop over each csv file
for csv_file in csv_files:
    
    # read the csv file
    df = pd.read_csv(csv_file)
    
    # filter for protein IDs with more than one protein and q-value < 0.01
    filtered_df = df[(df['Number of proteins'] > 1) & (df['Q-value'] < 0.01)]
    
    # get the number of rows in the filtered dataframe
    num_results = len(filtered_df)
    
    # get the filename without the directory path
    filename = os.path.basename(csv_file)
    
    # extract the method name and number of raw files from the filename
    method_name = filename.split("proteinGroups_")[1].split("_")[0:-1]
    method_name = "_".join(method_name)
    num_raw_files = filename.split("_")[-1].split(".")[0]
    
    # create a dictionary with the filename, method name, number of raw files, and number of results
    results_dict = {"Filename": filename, "Method": method_name, "Num_Raw_Files": num_raw_files, "Num_Results": num_results}
    
    # append the dictionary to the results list
    results_list.append(results_dict)
    
# convert the results list to a dataframe
results_df = pd.DataFrame(results_list)

# save the results to a csv file
results_df.to_csv("results.csv", index=False)
```
