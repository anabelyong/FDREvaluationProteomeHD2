```python
import csv

# Open the TSV file in read mode and CSV file in write mode
with open('subset_2000_decoy_FDR_001_psm.tsv', 'r') as tsvfile, open('subset_2000_decoy_FDR_001_psm.csv', 'w', newline='') as csvfile:
    # Create a TSV reader and a CSV writer
    tsvreader = csv.reader(tsvfile, delimiter='\t')
    csvwriter = csv.writer(csvfile, delimiter=',')
    
    # Loop through each row in the TSV file and write it to the CSV file
    for row in tsvreader:
        csvwriter.writerow(row)
        
print("Conversion and saving complete.")
```

```python
import pandas as pd
import re

# Define a function to extract the raw file name
def extract_raw_file_name(filename):
    if pd.isna(filename):
        return "unknown"
    else:
        return re.sub("\\..*", "", filename)

# Read in the CSV file
x = pd.read_csv('subset_750_decoy_FDR_001_psm.csv')

# Clean up the PSMId column
x['PSMId'] = x['PSMId'].str.strip().str.replace('\s+', ' ', regex=True)

# Extract the raw file name from the PSMId column
x['rawfile'] = x['PSMId'].apply(extract_raw_file_name)

# Group the DataFrame by the rawfile column and count the number of occurrences in each group
counts = x.groupby('rawfile').size()

# Print the counts
print(counts)
```
