```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df=pd.read_csv('/Users/anabelyong/miniconda3/lib/python3.10/site-packages/jupyterlab/Subset750RawfilesCut/Results/results_0.01_FDR_targets.csv')

# Group the DataFrame by method and raw_files
grouped = df.groupby(['Method', 'Num_Raw_Files']).agg({'Num_Results': 'sum'}).reset_index()

# Pivot the DataFrame to create separate lines for each method
pivoted = grouped.pivot(index='Num_Raw_Files', columns='Method', values='Num_Results')

# Plot the results using matplotlib
pivoted.plot.line(marker='o')
plt.xlabel('Number of Raw Files')
plt.ylabel('Number of Results')
plt.title('Results by Number of Raw Files')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas DataFrame
df=pd.read_csv('/Users/anabelyong/miniconda3/lib/python3.10/site-packages/jupyterlab/Subset750RawfilesCut/Results/results_0.01_FDR_targets.csv')

# Filter the DataFrame to include only the specified methods
selected_methods = ['savitski_no_remap', 'savitski_subset_no_remap', 'savitski_rescued_subset_no_remap']
filtered_df = df[df['Method'].isin(selected_methods)]

# Sort the DataFrame by number of raw files in ascending order
filtered_df = filtered_df.sort_values('Num_Raw_Files', ascending=True)

# Plot the results using matplotlib
for method in selected_methods:
    method_df = filtered_df[filtered_df['Method'] == method]
    plt.plot(method_df['Num_Raw_Files'], method_df['Num_Results'], label=method, marker='o')
plt.xlabel('# Raw Files')
plt.ylabel('# Target Protein Groups at 1% FDR')
plt.legend(loc='lower right')
plt.savefig('proteinGroups_savitski.png', bbox_inches='tight')
plt.show()


```
