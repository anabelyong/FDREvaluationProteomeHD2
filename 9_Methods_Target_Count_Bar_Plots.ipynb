```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

directory = '/Users/anabelyong/miniconda3/lib/python3.10/site-packages/jupyterlab/Subset750RawfilesCut/Bar_plots_Target_Decoy_Count/'

# Create an empty dataframe to store your results
results = pd.DataFrame()

# Loop through all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):  # assuming all files are in CSV format
        # Load the file into a Pandas dataframe
        df = pd.read_csv(os.path.join(directory, filename))
        
        # Filter the dataframe for proteins with Protein Type "sp" and Q-value < 0.01
        filtered_df = df[(df['Protein Type'] == 'sp') & (df['Q-value'] < 0.01)]
        
        # Count the number of proteins in the filtered dataframe
        protein_count = len(filtered_df)
        
        # Extract the method name from the filename
        method = filename.split("SeparateProteinIDs_")[1].split("_5000_RANDOM_")[0]
        
        # Add the filename, protein count, and method to the results dataframe
        results = results.append({'Filename': filename, 'Protein Count': protein_count, 'Method': method}, ignore_index=True)
    else:
        continue

# Calculate the average protein count for each method
averages = results.groupby(['Method']).mean()

# Calculate the standard deviation of protein count for each method
errors = results.groupby(['Method']).std()

# Create a bar plot with error bars and labels
fig, ax = plt.subplots()
ax.bar(averages.index, averages['Protein Count'], yerr=errors['Protein Count'], capsize=5)
ax.set_xlabel('Method')
ax.set_ylabel('Protein Count at 1% FDR')
ax.set_title('Protein Count by Method')
plt.xticks(rotation=90, ha='right')

# Add labels for the average protein count
for i, v in enumerate(averages['Protein Count']):
    ax.text(i, v, str(round(v, 2)), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()

# Save the image to a file
plt.savefig('/Users/anabelyong/miniconda3/lib/python3.10/site-packages/jupyterlab/Subset750RawfilesCut/Bar_plots_Target_Decoy_Count/protein_count_by_method.png', dpi=300)

# Show the plot
plt.show()
```
